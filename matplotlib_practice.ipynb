```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot.
#plt.bar(x_axis, y_axis, color="green", label='Boston')
plt.barh(x_axis, y_axis, color="green", label='Boston')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", marker="x", label = "Chicago")
plt.title("Pyber Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, marker="o",s = [i * 3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, color = "blue",s=[i*5 for i in y_axis], alpha=0.2)
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
explode_values = (0,0,0,0.3,0,0,0,0,0,0,0,0)
plt.pie(y_axis, labels = x_axis, explode=explode_values,autopct="%.1f%%")
plt.show()
```

```python
colors = ["slateblue", "magenta", "lightblue", "yellow", "yellowgreen", "greenyellow", "green", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(10,10))
explode_values = (0,0,0,0.3,0,0,0,0,0,0,0,0)
plt.pie(y_axis, labels = x_axis, explode=explode_values,autopct="%.2f%%",colors=colors,)
plt.show()
```

```python
stdev = statistics.stdev(y_axis)
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=5)
```

```python
from matplotlib.ticker import MultipleLocator
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51, step=5.0))
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
